### [Выделение пользовательских сессий](#start)
<a id="step4.1"></a>

Для того, чтобы точнее определить взаимосвязи и особенности в поведении пользователей, через события, которые они совершают, следует объединить эти события по какому-либо признаку. Один очевидный признак у нас уже есть - это пользователь. Другой признак, релевантный для мира мобильных приложений маркетплейсов - это пользовательская сессия. В рамках сессии можно проследить, как пользователь перемещался по цепочке событий и попыться найти логические закономерности.

Если бы мы пытались сопоставить два события одного пользователя без учета сессии, мы могли бы прийти к неверным выводам. Например, пытясь проанализировать событие просмотра обявлений в октябре 2019 и звонка продавцу в ноябре 2019. Скорее всего эти события вовсе бы относились к разным объявлениям.

К сожалению, у нас нет четкого критерия, который бы разделял сессии. Поэтому будем выделять сессии с помощью допущений. Для этого существует несколько подходов. 

Например, мы можем допустить среднюю продолжительность сессии и поделить весь датасет на это значение. Тогда все сессии пользователя (или даже всех пользователей) будут равны или меньше заданного параметра, что не совсем реалистично.

Или мы можем сделать допущение о признаке начала или окончания сессии. Таким признаком может быть временной промежуток между окончаением одной сессии и началом другой. Другими словами, тайм-аут между сессиями. Если мы сделаем допущение, что пользователь "отвлекается" от текущей сессии не более чем на заданное время, то сможем определить границы сессий через однородные события, между которыми прошло больше времени, чем заданный тайм-аут.

Для нашего проекта воспользуемся методом тайм-аута. Размер контрольного тайм-аута установим на уровне 30 мин.

Считаем, что пользователь начинает "работу" с приложением с поиска нужного объявления. Появление, просмотр карточек и другие события явялеются уже следствие каких-то результатов поиска.

```python
# сортируем датасет, чтобы результат функции .diff() дал корректный результат
mobile_dataset_sorted = mobile_dataset.sort_values(['user_id', 'event_time'])

# считаем временную разницу между событиями
datetime_diff = mobile_dataset_sorted.groupby('user_id')['event_time'].diff()

# создаем переменную с индексами и логическим значением признака начала новой сессии
new_session = (datetime_diff.isnull()) | (datetime_diff > timedelta(minutes=30))

# определяем порядковый номер каждой новой сессии для каждого пользователя
mobile_dataset_sorted['session_n'] = mobile_dataset_sorted.loc[new_session, ['user_id', 'event_time']] \
    .groupby('user_id').rank(method='first').astype(int)

# заменяем пропуски в НЕ первых событиях каждой сессии соответствующим номером сессии
mobile_dataset_sorted['session_n'] = mobile_dataset_sorted['session_n'].fillna(method='ffill').astype(int)

# создаем уникальный идентификатор каждой сессии
mobile_dataset_sorted['session_id'] = mobile_dataset_sorted['user_id'].str.cat(mobile_dataset_sorted['session_n'].values.astype(str))
```

### [Анализ метрик](#start)
<a id="step4.2"></a>

Выбирая метрики для анализа стоит еще раз вспомнить, что из себя представляет наш бизнес и наше мобильное приложение. 

Наша бизнес модель заключается в том, что мы продаем владельцам объвлений продвижение в поиске мобильного приложения. Владелецы объявлений получают выгоду, когда пользователи-покупатели обращают внимание на их объявление, переходят в карточку и контактируют с продавцом. Нам важно, чтобы продвижение работало эффективно и пользователи совершали как можно больше целевых событий - открывали объявления, а еще лучше, звонили продавцу.

Не менее важно, чтобы пользователи чаще заходили в приложение, искали товары и контактировали с продавцами. 

Чтобы оценить нашение приложение, рассчитаем несколько метрик с учетов вводных, описанных выше. Выберем следующие метрики:
1. Удержание (Retention Rate) - как насколько регулярно пользователи пользуются приложением
2. Время в приложении - как много времени пользователи проводят в приложении с учетом ограничения по тайм-ауту
3. Частота целевых действий - проанализируем как часто пользователи просматривают контакты в объяалениях
4. Конверсия в целевое действие - оценим сколько пользователей совершаю целевое действие в рамках сессии

Для расчете перечисленных показателей в датасете с событиями добавим дополнительные расчетные параметры.

```python
# определим начало первой сессии для каждого пользователя
mobile_dataset_sorted['first_session_start'] = \
    mobile_dataset_sorted.groupby('user_id')['event_time'].transform('min')

# создадим столбец с днём начала первой сессии для каждого пользователя
mobile_dataset_sorted['first_session_start_date'] = mobile_dataset_sorted['first_session_start'].dt.date

# определим начало каждой сессии для каждого пользователя    
mobile_dataset_sorted['session_start'] = \
    mobile_dataset_sorted.groupby(['user_id', 'session_n'])['event_time'].transform('min')

# определим конец каждой сессии для каждого пользователя    
mobile_dataset_sorted['session_end'] = \
    mobile_dataset_sorted.groupby(['user_id', 'session_n'])['event_time'].transform('max')

# рассчитаем продолжительность каждой сессии в минутах
mobile_dataset_sorted['session_duration_min'] = (mobile_dataset_sorted['session_end'] \
                                             - mobile_dataset_sorted['session_start']) \
                                                .dt.total_seconds().div(60).astype('int')
# display(mobile_dataset_sorted.head(20))
```

#### [Удержание](#start)
<a id="step4.2.1"></a>

Retention Rate, или коэффициент удержания, показывает, как долго клиенты остаются с компанией. Чтобы узнать Retention Rate, нужно разделить количество активных пользователей на текущий день на количество активных пользователей на первый день.

**Порядок расчета Retention Rate**

- Рассчитать лайфтайм пользователя для каждой сессии.
- Построить сводную таблицу удержания, где названия строк — это даты первого посещения пользователей, названия столбцов — лайфтайм, а значения в «ячейках» — количество уникальных идентификаторов пользователей.
- Вычислить размеры когорт и занести результаты в отдельную таблицу. За когорту примем пользователей с общим днём начала первой сессии.
- Объединить таблицы размеров когорт и удержания.
- Разделить каждую «ячейку» таблицы удержания на соответствующий размер когорты.

Также следует задать момент и горизонт анализа и учесть эти параметры при расчете коэффициента удержания. Для этого необходимо:
- Задать момент и горизонт анализа.
- Определить, игнорировать ли горизонт.
- Рассчитать максимально возможную дату привлечения пользователей.
- Исключить из анализа пользователей, пришедших после этой даты.
- Исключить из результата все лайфтаймы, превышающие горизонт анализа.

Шагов расчета получилось много, код будет объемным. Оформим его в виде функции.

```python
def get_retention(
    sessions, 
    observation_date, 
    horizon_days, 
    dimensions=[],
    ignore_horizon=False
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = sessions.query('first_session_start_date <= @last_suitable_acquisition_date')

    # считаем лайфтайм
    result_raw['lifetime'] = (
        result_raw['session_start'] - result_raw['first_session_start']
    ).dt.days

    # рассчитываем удержание
    result_grouped = result_raw.pivot_table(
        index=dimensions, columns='lifetime', values='user_id', aggfunc='nunique'
    )
    cohort_sizes = (
        result_raw.groupby(dimensions)
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    result_grouped = cohort_sizes.merge(
        result_grouped, on=dimensions, how='left'
    ).fillna(0)
    result_grouped = result_grouped.div(result_grouped['cohort_size'], axis=0)

    # исключаем все лайфтаймы, превышающие горизонт анализа
    result_grouped = result_grouped[
        ['cohort_size'] + list(range(horizon_days))
    ]

    # восстанавливаем столбец с размерами когорт
    result_grouped['cohort_size'] = cohort_sizes

    # возвращаем таблицу удержания
    return result_grouped
```

**Момент и горизонт анализа данных**

В качестве момента анализа данных используем последний день, за который имеются данные. Представим, что мы проводим анализ самых свежих данные, которые поступили к нам только что.

Горизон анализа зададим в размере 14 дней. Считаем, что хотя бы раз в две недели пользователь должен использовать приложение, чтобы не потерять интерес к нему на более длительном периоде. Не берем более короткий горизонт, потому что на нашей площадке продаются товары не первой необходимости, к тому же подержанные. Маловероятно, что даже активный пользователь будет каждую неделю искать себе что-то "ненужное".

```python
# зададим параметры функции
observation_date = mobile_dataset_sorted['event_time'].max().date()
horizon_days = 14


# вызовем функцию расчета коэффициента удержания
retention = get_retention(
    mobile_dataset_sorted, observation_date, horizon_days, dimensions=['first_session_start_date']
)

# построим хитмэп удержания
plt.figure(figsize = (15, 6))  # задаём размер графика
sns.heatmap(retention.drop(columns = ['cohort_size', 0]), # удаляем размеры когорт и удержание в первый день
            annot = True,  # включаем подписи
            fmt = '.2%'  # отображаем значения в виде процентов
           )
plt.ylabel('Дата первого события')
plt.title('Тепловая карта удержания')
plt.show()

# построим линейный график удержания
# report = retention.drop(columns=['cohort_size', 0])
# report.T.plot(grid=True, xticks=list(report.columns.values), figsize=(15, 5))
# plt.xlabel('Лайфтайм')
# plt.title('Кривые удержания с разбивкой по когортам')
# plt.show()
```

Других каких-либо персональных признаков пользователей нам не дано. В связи с этим попробуем разделить пользователей на основании событий, с которыми они сталкивали во время сессий.

Мы помним, что наш заказчик получает желаемое, когда пользователь переходит от просмотра рекомендаций к просмотру контактов в рекомендованных предложениях. ### [Выделение групп пользователей](#start)
<a id="step5.1"></a>

Еще раз сформулируем критерии разделения пользователей на группы:
- группа 1: пользователь испытал событие `tips_show` И одно или несколько событий поиска `map`, `search_1`-`search_7`
- группа 2: пользователь испытал одно или несколько событий поиска `map`, `search_1`-`search_7` И не испытал `tips_show`
- группа 3: все кто не попал в группу 1 или 2

```python
# построим сводную таблицу с количеством каждого события по каждому пользователю
events_by_user = mobile_dataset_sorted.pivot_table(
    index = 'user_id',
    columns = 'event_name',
    values = 'event_time',
    aggfunc = 'count')

# напишем функцию для определение группы пользователя
def user_groups(df):
    if (
        not pd.isnull(df['tips_show']) and (
            not pd.isnull(df['map']) or
            not pd.isnull(df['search_1']) or
            not pd.isnull(df['search_2']) or
            not pd.isnull(df['search_3']) or
            not pd.isnull(df['search_4']) or
            not pd.isnull(df['search_5']) or
            not pd.isnull(df['search_6']) or
            not pd.isnull(df['search_7'])
            )
        ): return 1
    elif (
        not pd.isnull(df['map']) or
        not pd.isnull(df['search_1']) or
        not pd.isnull(df['search_2']) or
        not pd.isnull(df['search_3']) or
        not pd.isnull(df['search_4']) or
        not pd.isnull(df['search_5']) or
        not pd.isnull(df['search_6']) or
        not pd.isnull(df['search_7'])
        ): return 2
    else: return 3

# применим функцию к сводной таблице и поместим номера групп в отдельную колонку
events_by_user['user_group'] = events_by_user.apply(user_groups, axis = 1)

# перенесем в первичный датасет группы пользователей из сводной таблицы
mobile_dataset_sorted = mobile_dataset_sorted.merge(events_by_user['user_group'], on='user_id', how='left')

# выведем количество пользователей в каждой группе
(mobile_dataset_sorted.groupby('user_group', as_index=False)
                     .agg(user_nunique=('user_id', 'nunique')))
```

### [Проверка гипотезы 1](#start)
<a id="step6.1"></a>

Проверка заданной гипотезы: *две группы пользователей, пришедшие в приложение из разных источников (`yandex` и `google`) демонстрируют разную конверсию в просмотры контактов.*  

H0 - конверсия пользователей из источника `yandex` РАВНО конверсия пользователей из источника `google`  
H1 - конверсия пользователей из источника `yandex` НЕ РАВНО конверсия пользователей из источника `google`

Для проверки гипотезы мы должны задать уровень значимости и выбрать метод.
Уровень значимости alpha зададим на уровне общепринятого значения 5%.
В качестве метода проверки выберем `scipy.stats.ttest_ind()`, так как нам необходимо сравнить между собой средние двух генеральных совокупностей на основании выборок из них. Одна совокупность - это пользователи из источника `yandex`, другая - пользователи из источника `google`.

```python
# объединим датасеты с пользователями и источниками
dataset = pd.merge(mobile_dataset_sorted, mobile_sourсes, on=['user_id'])

# создадим таблицу для теста с исходными данными по количеству пользователей, совершивших контрольное событие
events_for_tests = (
    dataset.pivot_table(
    index='event_name',
    columns = 'source',
    values = 'user_id',
    aggfunc = 'nunique')
).query('event_name == "contacts_show"').reset_index()

# оставим только контрольное событие
# events_for_tests = events_for_tests.query('event_name == "contacts_show"')

# создадим таблицу для теста с исходными данными по количеству участников в группах
groups_for_tests = (
    dataset.pivot_table(
    index = 'source',
    values = 'user_id',
    aggfunc = 'nunique'
    )
    .T
    .reset_index()
)

display(events_for_tests)
display(groups_for_tests)
```

Напишем функцию для проверки гипотез на различных исходных данных. Для функции будем задавать количество пользователей в группе `trials`, количество участников события `successes` и критический уровень статистической значимости `alpha`.

Далее используем функцию для проверки гипотезы.

```python
def exp_check(trials, successes, alpha):
    
    # переведем данные в формат array для работы с библиотекой numpy
    successes = np.array(successes)
    trials = np.array(trials)

    # рассчитаем пропорцию успехов в первой группе:
    p1 = successes[0]/trials[0]

    # рассчитаем пропорцию успехов во второй группе:
    p2 = successes[1]/trials[1]

    # рассчитаем пропорцию успехов в комбинированном датасете:
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

    # рассчитаем разницу пропорций в датасетах
    difference = p1 - p2 

    # рассчитаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

    # зададим стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = st.norm(0, 1) 

    # рассчитаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / mth.sqrt(
        p_combined * (1 - p_combined) * (1 / trials[0] + 1 / trials[1])
    )

    # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = st.norm(0, 1)
    
    # рассчитаем p-value
    p_value = (1 - distr.cdf(abs(z_value))) * 2

    # выведем p-value
    print('p-значение: ', p_value)

    # определим результаты проверки гипотезы
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print(
            'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
        )
```

```python
# зададим критический уровень статистической значимости
alpha = .05
```

```python
# зададим источники для проверки
sources = ['yandex', 'google']

# циклом выберем нужные значения, подставим их в функцию и запустим её 
for i in events_for_tests.index:
    events = events_for_tests.loc[0, sources]
    groups = groups_for_tests.loc[0, sources]
#     print(events_for_tests.loc[i, 'event_name'])
#     print(f'\nCобытия\n{events}\n')
#     print(f'Группы\n{groups}\n')
    exp_check(groups, events, alpha)
    print(f'\n')

# рассчитаем конверсию в целевое действие в разрезе групп
coversion = events_for_tests.loc[0, 'yandex'] / groups_for_tests.loc[0, 'yandex']
print(f'Конверсия в целевое действие в разрезе пользователей из источника yandex: {coversion:.2f}')

coversion = events_for_tests.loc[0, 'google'] / groups_for_tests.loc[0, 'google']
print(f'Конверсия в целевое действие в разрезе пользователей из источника google: {coversion:.2f}')
```
